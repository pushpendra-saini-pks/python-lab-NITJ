```python
# -----------------------------
# STRING OPERATIONS
# -----------------------------

# Concatenation
print("good" + " morning")
print("python" + "technology")
print("python" + "3.5")
print(1234)
```

```python
# -----------------------------
# STRING MULTIPLICATION
# -----------------------------
print("Good" * 5)
print(5 * "Good")
```

```python
# -----------------------------
# STRING SLICING
# -----------------------------
s = "python is Easy!!!"
print(s)
print(s[0])        # first char
print(s[3:9])      # slice
print(s[4:])       # from index 4 to end
print(s[-1])       # last char
print(s[:5])       # first 5 chars
```

```python
# -----------------------------
# UNICODE / ESCAPE SEQUENCES
# -----------------------------
print("what's your name?")
print('what\'s your name?')
print("The boy replied, \"My name is Aaditya\"")
print("The boy replied,\n\"My name is Aaditya\"")
print("The boy replied,\t\"My name is Aaditya\"")

# This line was incorrect — fixed by escaping the backslash:
print("The boy replied \\ My name is Aaditya")

print("*****************************************")
```

```python
# Escape sequences
print("\\")
print("\'")
print("\"")
print("Hello\nworld")
print("Hello\tworld")
print("\056")      # octal
print("\x2E")      # hex
print("\x87")      # hex
```

```python
# NUMBER CONVERSIONS
# -----------------------------
print(int(0b11011))   # binary to int
print(oct(51))        # int to octal
print(int(0x360))     # hex to int
print(hex(51))        # int to hex
print(int(0x33))      # hex to int
```

```python
# IDENTITY OPERATORS
# -----------------------------
x = [1, 2]
y = [1, 1]
print("x is y:", x is y)

z = x
print("x is z:", x is z)

x1 = 5
y1 = 5
x2 = "Hello"
y2 = "Hello"
x3 = [1, 2, 3]
y3 = [1, 2, 3]

print(x1 is not y1)
print(x2 is y2)
print(x3 is y3)
```

```python
# MEMBERSHIP OPERATORS
# -----------------------------
message = "hello world"
dict_data = {1: 'a', 2: 'b'}

print('h' in message)
print('hello' in message)
print('HELLO' in message)

print(1 in dict_data.keys())
print('a' in dict_data.values())
```

```python
# TERNARY OPERATOR
# -----------------------------
x = -0.5
result = "positive" if x > 0 else "negative"
print(result)

x = 5
result = "positive" if x > 0 else "negative"
print(result)
```

```python
# EXPRESSIONS
# -----------------------------
expr = 10 + 20 * 30
print(expr)

print((40 + 20) * 30 / 10)
print(40 + (20 * 30) / 10)

print(False == False or True)
print(False == (False or True))
```

```python
# USER INPUT
# -----------------------------
name = input("What's your name? ")
age = input("Enter your age: ")
print(name + " you are " + age + " years old")

```

```python
# Basic use of format()
name = "Aaditya"
age = 20

print("My name is {} and I am {} years old.".format(name, age))
```

```python
# Using index numbers inside format()
print("I live in {1} and my friend lives in {0}".format("Delhi", "Jaipur"))
```

```python
# Using named placeholders
print("Student: {name}, Roll No: {roll}".format(name="Pushpendra", roll=25201312))
```

```python
# Formatting numbers
pi = 3.1415926
print("Value of pi = {:.2f}".format(pi))
```

```python
text = "Hello"

print("Left aligned  :", "{:<10}".format(text))
print("Right aligned :", "{:>10}".format(text))
print("Centered      :", "{:^10}".format(text))
```

```python
num = 25

print("Binary: {0:b}".format(num))
print("Octal: {0:o}".format(num))
print("Hexadecimal: {0:x}".format(num))
```

```python
# Creating a simple table
print("{:<10} {:<10} {:<10}".format("Name", "Sub", "Marks"))
print("{:<10} {:<10} {:<10}".format("Jatin", "Math", 95))
print("{:<10} {:<10} {:<10}".format("Riya", "Science", 88))
print("{:<10} {:<10} {:<10}".format("Aman", "English", 76))
```

```python
marks = 430
total = 500
percentage = (marks / total) * 100

print("Your result is {:.2f}%".format(percentage))
```
