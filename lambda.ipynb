```python
def fact(n):
    if n==0 or n==1:
        return n
    return n*fact(n-1)


x=fact(5)
print(x)
```

```python
# sum using recursion 1 to n
def f(n):
    if n==0:
        return 0
    return n+f(n-1)

n=10
print(f(n))
```

```python
def m(arr,n):
    if n==1:
        return arr[0]
    # return min (arr[n-1],m(arr,n-1))
    if arr[n-1] > arr[n-2]:
        return m(arr,n-2)

arr=[1,4,24,17,-5,10,-22]
n=len(arr)
print (m(arr,n))
```

```python
# write  a program a fabbonacci series using recursion
def fab(n):
    if n<=1:
        return n
    return fab(n-1)+fab(n-2)    
n=10
for i in range(n):
    print(fab(i))   
```

```python
# write a program to reverse a string
def revs(s):
    if len(s)==0:
        return s
    else:
        return s[-1]+revs(s[:-1])
s="Pushpendra"
print(revs(s))
```

```python
# calculate exponetial using recursion
def exp(a,b):
    if b==0:
        return 1
    return a*exp(a,b-1)
print(exp(2,5)) 
```
