```python
# 1. Increment number if it is positive
number = int(input("Please enter a number: "))
if number > 0:
    number += 1
print("Updated number:", number)
```

```python
# 2. Check eligibility for vote
age = int(input("\nEnter age: "))
if age >= 18:
    print("You are eligible to vote")
else:
    print(f"You are {age} NOT eligible to vote\nYou must wait", 18 - age, "more years")
```

```python
# 3. Determine character type
char = input("\nPress any key: ")
if char.isalpha():
    print(f" {char} is a character")
elif char.isdigit():
    print(f" {char} is a digit")
elif char.isspace():
    print(f" {char} is a white space")
else:
    print(f" {char} is a Special Character")
```

```python
# 4. Max of two numbers
a, b = map(int, input("Enter two numbers: ").split())
print(f"Maximum of {a} and {b} is ", max(a, b))
```

```python
# 5. Convert lowercase â†” uppercase
char = input("\nEnter an alphabet: ")
if char.islower():
    print(f"Uppercase of {char} is ", char.upper())
elif char.isupper():
    print(f"Lowercase of {char} is ", char.lower())
else:
    print("Not an alphabet")
```

```python
# 6. Check number is +, -, or zero
num = int(input("\nEnter a number: "))
if num > 0:
    print(f"the num {num} is Positive number")
elif num < 0:
    print(f" the num {num} is Negative number")
else:
    print("Zero")
```

```python
# 7. Check vowel or not
char = input("\nEnter a character to check vowel: ")
if (char == 'a' or char == 'e' or char == 'i' or char == 'o' or char == 'u' or
    char == 'A' or char == 'E' or char == 'I' or char == 'O' or char == 'U'):
    print(f" {char} is Vowel")
else:
    print(f" {char} is Not vowel")
```

```python
# 9. Check whether input is character or number, uppercase/lowercase
char = input("\nEnter something: ")

if '0' <= char <= '9':
    print(f" {char} is Digit")
elif 'a' <= char <= 'z':
    print(f" {char} Lowercase character")
elif 'A' <= char <= 'Z':
    print(f" {char} Uppercase character")
else:
    print(f"{char} Special character")
```

```python
# 10. Largest among three numbers (NO max() function)
a, b, c = map(int, input("\nEnter three numbers: ").split())

if a > b and a > c:
    print(f"Largest among {a},{b},{c} is ", a)
elif b > a and b > c:
    print(f"Largest among {a},{b},{c} is ", b)
else:
    print(f"Largest among {a},{b},{c} is ", c)
```

```python
# 11. Armstrong number
num = int(input("\nEnter number to check Armstrong: "))
temp = num
digit = 0

# count digits
n = num
while n > 0:
    digit += 1
    n //= 10
    
# calculate sum of digits^digit
n = num
arm = 0
while n > 0:
    rem = n % 10
    arm += rem ** digit
    n //= 10

if arm == temp:
    print(f"{num} is a Armstrong number")
else:
    print(f" {num} is Not an Armstrong number")
```

```python
# 12. Reverse a number
num = int(input("\nEnter number to reverse: "))
temp = num
rev = 0
while num > 0:
    rem = num % 10
    rev = rev * 10 + rem
    num //= 10
print(f"Reverse of {temp} is ", rev)
```

```python
# 13. Factorial
num = int(input("\nEnter number for factorial: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print(f"Factorial of {num} is", fact)
```

```python
# 14. Multiplication table
num = int(input("\nEnter number for multiplication table: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```
