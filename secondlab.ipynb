```python
import math
```

```python
# 1. Number conversion and square root
number = int(input("Enter a number: "))
print("Hex form of number:", hex(number))
print("Octal form of number:", oct(number))
print("Square root of number:", math.sqrt(number))
```

```python
# 2. Area of triangle (Heron's formula)
a = float(input("Enter length of side a: "))
b = float(input("Enter length of side b: "))
c = float(input("Enter length of side c: "))
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area of triangle:", area)
```

```python
# 3. Distance between two points (x1,y1) & (x2,y2)
x1 = float(input("Enter x-coordinate of point 1: "))
y1 = float(input("Enter y-coordinate of point 1: "))
x2 = float(input("Enter x-coordinate of point 2: "))
y2 = float(input("Enter y-coordinate of point 2: "))
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print(f"Distance between the ({x1},{y1}) and ({x2},{y2})", distance)
```

```python
# 4. Basic arithmetic operations
a = float(input("Enter number 1: "))
b = float(input("Enter number 2: "))

print(f"Addition of {a} and {b} is ", a + b)
print(f"Subtraction of {a} and {b} is ", a - b)
print(f"Division of {a} and {b} is ", a / b)
print(f"Multiplication of {a} and {b} is ", a * b)  
```

```python
# 5. Bill amount calculation
print("*" * 54)
print("******************* BILL *******************")
print("*" * 54)

quantity = float(input("Quantity sold: "))
price = float(input("Price per item: "))
amount = quantity * price
print("Amount:", amount)

discount = float(input("Enter discount: "))
final_amount = amount - discount
print("Final amount after discount:", final_amount)
```

```python
# 6. Fahrenheit to Celsius
fahrenheit = float(input("Enter temperature in Fahrenheit: "))
celsius = (fahrenheit - 32) * 5 / 9
print(f"Temperature in Celsius: {celsius}")
# 7. Celsius to Fahrenheit
celsius = float(input("Enter temperature in Celsius: "))    
fahrenheit = (celsius * 9 / 5) + 32
print(f"Temperature in Fahrenheit: {fahrenheit}")
```

```python
# 7. Piggy Bank Calculation
num_coins_1 = int(input("Enter number of 1-rupee coins: "))
num_coins_2 = int(input("Enter number of 2-rupee coins: "))
num_coins_5 = int(input("Enter number of 5-rupee coins: "))
total_amount = (num_coins_1 * 1) + (num_coins_2 * 2) + (num_coins_5 * 5)
print("Total amount in piggy bank:", total_amount)  
```
