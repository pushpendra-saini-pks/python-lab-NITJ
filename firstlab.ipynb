```python
## Addition, Subtraction and Multiplication of two numbers
a = 13
b = 6

c = a + b
d = a - b
e = a * b

print("sum =", c)
print("substraction", d)
print("multiplication", e)
```

```python
## Power of a number
a = 10
b = 2
print(f"{a} ki power {b} is =  {a**b}")
```

```python
## Different types of division
a = 7
b = 2
print(f"divide : {a/b}")
print(f"floor division : {a//b}")
print(f"negative floor division : {(-a//b)}")
```

```python
# a//b rounds results towards the whole number and (-a//b) rounds the results towards the nagative infinite

print("negative power of any number :", 10**(-1))
print(2**(-1))
print(7**(0))
print((-7)**(0))
```

```python
# comparision operators
x = 3
y = 2
print(x == y)

x = 3
y = 3
print(x == y)
print(x < y)
print(x > y)
print(x != y)
print(x <= y)
print(x >= y)
```

```python
# assignment operators
x = 5

x += 3
print("Increment operator (+= 3):", x)

x *= 3
print("Multiplicative increment operator (*= 3):", x)

x -= 3
print("Decrement operator (-= 3):", x)

x %= 3
print("Modulo decrement operator (%= 3):", x)

x /= 3
print("Division operator (/= 3):", x)

x //= 3
print("Floor division operator (//= 3):", x)

x **= 3
print("Exponentiation operator (**= 3):", x)

```

```python
# comparision operators
is_allowence = False
print(not is_allowence)

a = 5
b = 6
print(a < 5 and b > 19)
print(a < 5 or b > 19)
```
