```python
#WAP that creates a list of numbers from 1-20 that are divisible by 2 or 4
def divisible_by_2_or_4(num):
    if num%2==0 or num%4==0:
        return num
    
l1 = list(range(1,21))
l2 = list(filter(divisible_by_2_or_4,l1))
print(l2)
```

```python
#WAP that creates a list of sqaure of numbers from 1-10 ....
from functools import reduce
squares = list(map(lambda x: x**2, range(1, 11)))

sum_of_squares = reduce(lambda a, b: a + b, squares)
print("List of squares:", squares)
print("Sum of squares:", sum_of_squares)
```

```python
#WAP to check whether the given country is a member of BRICS or not
brics = ["Brazil", "Russia", "India", "China", "South Africa"]

country = input("Enter the name of a country: ").strip().title()

result = list(filter(lambda x: x == country, brics))

if result:
    print(country, "is a member of BRICS.")
else:
    print(country, "is NOT a member of BRICS.")
```

```python
#Q1: WAP that creates a list of 10 random integers, then create two lists — odd and even

import random
numbers = [random.randint(1, 100) for _ in range(10)]
print("Original list:", numbers)

odd_list = list(filter(lambda x: x % 2 != 0, numbers))
even_list = list(filter(lambda x: x % 2 == 0, numbers))

print("Odd list:", odd_list)
print("Even list:", even_list)
```

```python
# Q2: Convert a string of uppercase letters into lowercase using map()
text = "HELLO WORLD"

lower_text = ''.join(map(lambda ch: ch.lower(), text))

print("Lowercase string:", lower_text)
```

```python
# Q3: Find the largest value in a list using reduce()
from functools import reduce

numbers = [10, 25, 8, 99, 45, 67]

largest = reduce(lambda a, b: a if a > b else b, numbers)
print("Largest number:", largest)
```

```python
# Q4: Scan an email address and form a tuple of (username, domain)
email = input("Enter your email address: ")

if '@' in email:
    username, domain = email.split('@')
    print("Tuple:", (username, domain))
else:
    print("Invalid email address")
```

```python
# Q5: List with positive and negative numbers → make a tuple of only positive values
numbers = [-5, 3, -2, 7, -9, 10, 0]

positive_tuple = tuple(filter(lambda x: x > 0, numbers))

print("Tuple of positive numbers:", positive_tuple)
```

```python
# Q6: Find intersection, union, and symmetric difference between two sets (Coders & Analysts)
coders = {"Alice", "Bob", "Charlie", "David"}
analysts = {"Charlie", "Eve", "Frank", "Bob"}

# a) People working as Coders as well as Analysts
both = coders & analysts
print("a) Coders and Analysts both:", both)

# b) People working as Coders or Analysts
either = coders | analysts
print("b) Coders or Analysts:", either)

# c) People working as Analysts but not Coders
only_analysts = analysts - coders
print("c) Analysts but not Coders:", only_analysts)

# d) People working in only one of the groups
only_one = coders ^ analysts
print("d) Working in only one group:", only_one)
```
