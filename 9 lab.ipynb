```python
## Q1. Write a program that has classes such as Student, Course and Department. Enroll a student in a course of a particular department.
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = []

    def add_course(self, course):
        self.courses.append(course)


class Course:
    def __init__(self, course_name, dept):
        self.course_name = course_name
        self.department = dept
        self.students = []
        dept.add_course(self)

    def enroll(self, student):
        self.students.append(student)


class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    def enroll_in_course(self, course):
        course.enroll(self)


# ----------- Testing --------------
dept1 = Department("Computer Science")
course1 = Course("Data Structures", dept1)

stud1 = Student("Rahul", 101)
stud1.enroll_in_course(course1)

print("Department:", dept1.name)
print("Course:", course1.course_name)
print("Enrolled Students:", [s.name for s in course1.students])
```

```python
## Q2. Write a program with class Bill. The users have the option to pay the bill either by cheque or by cash. Use the inheritance to model this situation.
class Bill:
    def __init__(self, amount):
        self.amount = amount

    def pay(self):
        print("Payment method not specified.")


class ChequePayment(Bill):
    def __init__(self, amount, cheque_no):
        super().__init__(amount)
        self.cheque_no = cheque_no

    def pay(self):
        print(f"Paid ₹{self.amount} by Cheque No: {self.cheque_no}")


class CashPayment(Bill):
    def pay(self):
        print(f"Paid ₹{self.amount} in Cash")


# ------------ Testing --------------
b1 = ChequePayment(2500, "CHQ1456")
b1.pay()

b2 = CashPayment(1800)
b2.pay()
 
```

```python
## Q3.Write a program to overload the - =operator to subtract two Distance Objects.
class Distance:
    def __init__(self, meters):
        self.m = meters

    def __isub__(self, other):
        self.m = self.m - other.m
        return self

    def show(self):
        print(f"Distance: {self.m} meters")


# ----------- Testing --------------
d1 = Distance(50)
d2 = Distance(30)

d1 -= d2   # calls __isub__
d1.show()
```

```python
## Q4. Write a program with two classes for calculating the distance. On has distance specified in meters and other has distance in kilometres. These functions take argument of class Distance that converts the distance into kilometres and meters.
class MeterDistance:
    def __init__(self, meters):
        self.m = meters

    def to_km(self):
        return self.m / 1000


class KilometerDistance:
    def __init__(self, km):
        self.km = km

    def to_m(self):
        return self.km * 1000


# ----------- Testing --------------
m1 = MeterDistance(1500)
k_from_m = m1.to_km()
print("1500 meters =", k_from_m, "km")

k1 = KilometerDistance(3)
m_from_k = k1.to_m()
print("3 km =", m_from_k, "meters")
```
